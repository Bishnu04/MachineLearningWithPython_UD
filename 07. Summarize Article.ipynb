## Pre Process Downloaded Text
Read the Article from a file/web
List of words that are present in the list  
From above list remove stopwords  
Filter-out stopwords from our article  


```python
#Read the article and store it in a variable
article = ""
with open('Data/NLP/article.txt', 'r') as file:
    data = file.read()
article = data
print(article)
```

```python
#Preprocess Downloaded Text
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation

text = article
sents = sent_tokenize(text)

#Convert article into list of words
word_sent = word_tokenize(text.lower())

#Remove stopwords 
customstopwords=set(stopwords.words('english')+list(punctuation))

#Take all words after removing stopwords from it
wordslist=[word for word in word_sent if word not in customstopwords]
wordslist
```

```python
#Extracting the most common words
from nltk.probability import FreqDist
#Construct a frequency distribution of words
freq = FreqDist(wordslist)
print(freq.most_common)
```

```python
#find top 10 common words in the article
from heapq import nlargest

print (nlargest(10, freq, key=freq.get))

#Compute the score for each sentence (Keys -> Sentences, Values -> Significance Scores)
from collections import defaultdict
#if a key is not present defaultdict will not throw error, it will add the key in the dict

ranking = defaultdict(int)

for i,sent in enumerate(sents):
    for w in word_tokenize(sent.lower()):
        if w in freq:
            ranking[i] += freq[w]

print (ranking)

#this will not store in sorted order
top_sent = nlargest(3, ranking, key=ranking.get)
print (top_sent)

#sort and print in ascending order
print ([sents[j] for j in sorted(top_sent)])
```

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation
from nltk.probability import FreqDist
from heapq import nlargest
from collections import defaultdict

def summarize_article(text,top_n):
    sents = sent_tokenize(text)
    word_sent = word_tokenize(text.lower())
    customstopwords=set(stopwords.words('english')+list(punctuation))
    wordslist=[word for word in word_sent if word not in customstopwords]
    freq = FreqDist(wordslist)

    ranking = defaultdict(int)

    for i,sent in enumerate(sents):
        for w in word_tokenize(sent.lower()):
            if w in freq:
                ranking[i] += freq[w]
    top_sent = nlargest(top_n, ranking, key=ranking.get)
    print ([sents[j] for j in sorted(top_sent)])



text = ""
with open('Data/NLP/article.txt', 'r') as file:
    text = file.read()
summarize_article(text, 3)
```
